```python
import visual_automata
from visual_automata.fa.dfa import VisualDFA

s = eval(input("Enter the states: "))
sy = eval(input("Enter input symbols: "))
t = {}
for i in s:
    input1 = eval(input("Enter the transitions in the form of a dictionary:"))
    if input1 == "":
        continue
    t[str(i)] = input1
initial_si = eval(input("Enter initial state: "))
final_si = eval(input("Enter final state(s): "))

dfa = VisualDFA(
    states=s,
    input_symbols=sy,
    transitions=t,
    initial_state=initial_si,
    final_states=final_si,
)

dfa = VisualDFA(dfa)
print(dfa.table)
dfa.show_diagram()
```
